Merge dataframes

```python
# Step 1: Merge users_df with course_progress_df on employee_id
merged_df = pd.merge(users_df, course_progress_df, left_on='user_id', right_on='employee_id', how='left')

# Step 2: Merge merged_df with courses_df to include course details
merged_df = pd.merge(merged_df, courses_df, on='course_id', how='left')

# Step 3: Merge module_progress_df with course_modules_df to get module details for each course
module_df = pd.merge(module_progress_df, course_modules_df, on='module_id', how='left')

# Step 4: Merge the module_df with merged_df to incorporate module completion info
final_df = pd.merge(merged_df, module_df,left_on=['user_id', 'course_id'], right_on=['employee_id', 'course_id_x'], how='left')
```

```python
final_df = final_df[[
    'user_id', 'name', 'email', 'designation',   # From users_df
    'course_id', 'title', 'tag', 'completion_percentage', 'duration', 'score', 'completed_at_x',  # From course_progress_df and courses_df
    'module_id', 'module_title', 'is_completed', 'module_duration', 'completed_at_y'  # From module_progress_df and course_modules_df
]]

# Rename columns for clarity
final_df.rename(columns={
    'completed_at_x': 'course_completed_at',
    'completed_at_y': 'module_completed_at',
    'title': 'course_title',
    'tag': 'course_tag',
    'score': 'course_score',
    'duration': 'course_duration',
}, inplace=True)
```

Employee Course KPI: This captures each employeeâ€™s course progress (average completion percentage and performance score). This will be useful for understanding how far employees have progressed in their courses and how theyâ€™re performing.

```python
# Group by user and course to calculate total modules and completed modules per course
module_kpi = final_df.groupby(['user_id', 'course_id']).agg(
    total_modules=('module_id', 'count'),        # Total number of modules for the course
    modules_completed=('is_completed', 'sum')   # Number of completed modules (is_completed is 1 for completed)
).reset_index()

module_kpi.columns

display(module_kpi)
```

```python
# Merge the module_kpi with the final_df to incorporate total and completed modules
final_df = pd.merge(final_df, module_kpi, on=['user_id', 'course_id'], how='left')

display(final_df.columns)
```

```python
# Calculate completed percentage for the modules
final_df['completion_percentage'] = (final_df['modules_completed'] / final_df['total_modules']) * 100
```

```python
# Employee KPI: Each employee's course progress with total modules, completed modules, and module completion percentage
employee_course_kpi = final_df.groupby(['user_id', 'name', 'designation', 'course_id', 'course_title', 'course_tag']).agg(
    total_modules=('total_modules', 'first'),
    modules_completed=('modules_completed', 'first'),
    completion_percentage=('completion_percentage', 'first'),
    course_score=('course_score', 'first')
).reset_index()
```

```python
def calculate_performance_score(progress_df):
    # Group by employee and calculate total score and total max score
    performance_summary = progress_df.groupby('user_id').agg(
        total_score=('course_score', 'sum'),  # Sum of course scores
        total_max_score=('course_score', lambda x: len(x) * 100),  # Assuming max score for each course is 100
        completed_courses=('completion_percentage', lambda x: (x == 100).sum())  # Count of fully completed courses
    ).reset_index()

    # Calculate performance score as a percentage
    performance_summary['performance_score'] = (performance_summary['total_score'] / performance_summary['total_max_score']) * 100
    return performance_summary

# %%
# Update performance score for each employee
performance_summary = calculate_performance_score(final_df)
```

```python
final_df = pd.merge(final_df, performance_summary[['user_id', 'performance_score']], on='user_id', how='left')
```

Get unassigned courses of employees

```python
employee_course_combinations = pd.merge(users_df[['user_id', 'name', 'designation']],
                                        courses_df[['course_id', 'title', 'tag', 'duration', 'modules']],
                                        how='cross')  # Cartesian product

# Merge employee-course combinations with assigned courses to check what's already assigned
assigned_courses = pd.merge(course_assignment_df[['employee_id', 'course_id']],
                            employee_course_combinations,
                            left_on=['employee_id', 'course_id'],
                            right_on=['user_id', 'course_id'],
                            how='right')

# Filter for rows where the course is not yet assigned (i.e., NaN in assignment)
unassigned_courses = assigned_courses[assigned_courses['employee_id'].isna()]

# Drop irrelevant columns
unassigned_courses = unassigned_courses[['user_id', 'name', 'designation', 'course_id', 'title', 'tag', 'duration', 'modules']]

# Save the result as a CSV file
unassigned_courses.to_csv('data_warehouse/report/unassigned_courses.csv', index=False)
```
