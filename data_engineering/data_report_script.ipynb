```python
import pandas as pd
```

```python
users_df = pd.read_csv('data_warehouse/staging/users_staging.csv')
course_progress_df = pd.read_csv('data_warehouse/staging/course_progress_staging.csv')
module_progress_df = pd.read_csv('data_warehouse/staging/module_progress_staging.csv')
courses_df = pd.read_csv('data_warehouse/staging/courses_staging.csv')
course_modules_df = pd.read_csv('data_warehouse/staging/modules_staging.csv')
```

Merge dataframes

```python
# Step 1: Merge users_df with course_progress_df on user_id and employee_id
merged_df = pd.merge(users_df, course_progress_df, left_on='user_id', right_on='employee_id', how='left')

# Step 2: Merge courses_df with the previous result on course_id
merged_df = pd.merge(merged_df, courses_df, left_on='course_id', right_on='course_id', how='left')

# Step 3: Merge module_progress_df with course_modules_df on module_id and then join with the main dataframe
module_df = pd.merge(module_progress_df, course_modules_df, left_on='module_id', right_on='module_id', how='left')
final_df = pd.merge(merged_df, module_df, left_on=['user_id', 'course_id'], right_on=['employee_id', 'course_id'], how='left')

# Step 4: Drop duplicate columns and keep only relevant ones
final_df = final_df[[
    'user_id', 'name', 'email', 'designation', 'performance_score',  # From users_df
    'course_id', 'title', 'tag', 'completion_percentage', 'duration', 'completed_at_x',  # From course_progress_df and courses_df
    'module_id', 'module_title', 'is_completed', 'module_duration', 'completed_at_y'  # From module_progress_df and course_modules_df
]]

# Rename columns for clarity
final_df.rename(columns={
    'completed_at_x': 'course_completed_at',
    'completed_at_y': 'module_completed_at'
}, inplace=True)
```
